**Predicting if a person would buy life insurnace based on his age using logistic regression**

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
df=pd.read_csv(r"C:\Users\ASUS\NLP\insurance_data.csv")
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

**Preprocessing: Scale the data so that both age and affordibility are in same scaling range**

```python
## scaling the age by dividing with 100
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

```python
X_train_scaled
```

***Model Building: First build a model in keras/tensorflow and see what weights and bias values it comes up with.***

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])### input_shape=2 as for age and affordibility
## also initializing the weights to be One and Bias to be Zero

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=500)
```

```python
model.evaluate(X_test_scaled,y_test)
```
